```python
import pandas as pd

import plotly.graph_objects as go
from plotly.subplots import make_subplots
df1 = pd.read_csv(r"C:\Users\DELL\Downloads\AAPL.csv")
df2 = pd.read_csv(r"C:\Users\DELL\Downloads\SP500INDEX.csv")
merged = pd.concat([df1,df2],ignore_index=True)

merged['Stock'] = 'null'
merged.loc[:252,'Stock'] = 'APPL'
merged.loc[253:506,'Stock'] = 'S&P500'
merged.drop(columns = ['Adj Close','Open','High','Low','Close'],axis=1,inplace = True)
merged.groupby('Stock').mean()
```

### Some future visualization

```python
import pandas as pd

import plotly.graph_objects as go
from plotly.subplots import make_subplots
df1 = pd.read_csv(r"C:\Users\DELL\Downloads\AAPL.csv")
df2 = pd.read_csv(r"C:\Users\DELL\Downloads\SP500INDEX.csv")
df1['Change'] = (((df1['Open']-df1['Close'])/ df1['Open']) * 100)
df2['Change'] = (((df1['Open']-df1['Close'])/ df1['Open']) * 100)
df1.drop(columns = ['Adj Close'],axis=1,inplace = True)
df2.drop(columns = ['Adj Close'],axis=1,inplace = True)
# df1.query('Change > 1',inplace = True)
fig = make_subplots(specs = [[{"secondary_y":True}]])
fig.add_trace(go.Scatter(x = df1['Date'],y = df1["Open"],name = 'APPL'),secondary_y=True)
fig.add_trace(go.Scatter(x = df2['Date'],y = df2["Open"],name = 'S&P500'))
fig.show()
```
