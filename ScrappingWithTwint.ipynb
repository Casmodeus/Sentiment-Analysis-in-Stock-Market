```python
# Import Library
import twint
```

```python
# Instantiate and configure the twint-object
c = twint.Config()
c.Store_object = True
c.Pandas =True
c.Search = "$FB min_replies:0 min_faves:2 -filter:retweets"
c.Limit = 20000
c.Lang = 'en'
c.Geo = True
c.Since = '2020-07-31'
c.Until = '2020-11-05'
```

```python
import nest_asyncio
nest_asyncio.apply()
```

```python
# Run search
twint.run.Search(c)
```

```python
# Quick check
twint.storage.panda.Tweets_df.tail(100)
```

```python
# Cleanup
tweets = twint.storage.panda.Tweets_df.drop_duplicates(subset=['id'])
```

```python
# Reindex
tweets.index = range(len(tweets))
```

```python
# Remove non-english
from whatthelang import WhatTheLang
wtl = WhatTheLang()
```

```python
# This function makes easy to handle exceptions (e.g. no text where text should be)
# not really needed but can be useful 

def detect_lang(text):
    try: 
        return wtl.predict_lang(text)
    except Exception:
        return 'exp'
```

```python
# Added performance measure here...you can leave teh %%time line out

%%time

tweets['lang'] = tweets['tweet'].map(lambda t: detect_lang(t))
```

```python
# keep only english

tweets = tweets[tweets.lang == 'en']
```

```python
tweets.Store_csv = True
```
